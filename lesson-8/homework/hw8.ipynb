```python
# 4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.
def get_two_numbers():
    num1 = input("Enter the first number: ")
    num2 = input("Enter the second number: ")

    # Check if inputs are numeric
    if not (num1.replace('.', '', 1).isdigit() and num2.replace('.', '', 1).isdigit()):
        raise TypeError("Invalid input! Both values must be numbers.")

    # Convert to float
    num1 = float(num1)
    num2 = float(num2)

    print(f"You entered: {num1} and {num2}")
    return num1, num2

# Example usage
try:
    a, b = get_two_numbers()
except TypeError as e:
    print(e)
```

```python
# 12. Write a Python program to read first n lines of a file.
def read_first_n_lines(file_name, n):
    try:
        with open(file_name, 'r') as file:
            for i in range(n):
                line = file.readline()
                if not line:
                    break  # Stop if file has fewer than n lines
                print(line, end='')  # Avoid adding extra newline
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")

# Example usage
file_name = input("Enter the file name: ")
n = int(input("Enter the number of lines to read: "))
read_first_n_lines(file_name, n)
```

```python
# 13. Write a Python program to append text to a file and display the text.
def append_and_display(file_name, text_to_append):
    try:
        # Append text to the file
        with open(file_name, 'a') as file:
            file.write(text_to_append + '\n')

        # Read and display the updated content
        with open(file_name, 'r') as file:
            content = file.read()
            print("Updated file content:")
            print(content)
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")

# Example usage
file_name = input("Enter the file name: ")
text = input("Enter text to append: ")
append_and_display(file_name, text)

```

```python
# 14. Write a Python program to read last n lines of a file.
from collections import deque

def read_last_n_lines(file_name, n):
    try:
        with open(file_name, 'r') as file:
            last_lines = deque(file, maxlen=n)
        print("".join(last_lines))
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")

# Example usage
file_name = input("Enter the file name: ")
n = int(input("Enter the number of lines to read from the end: "))
read_last_n_lines(file_name, n)
```

```python
# 15. Write a Python program to read a file line by line and store it into a list.
def read_file_to_list(file_name):
    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()
            # Remove newline characters
            lines = [line.strip() for line in lines]
        return lines
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
        return []
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")
        return []

# Example usage
file_name = input("Enter the file name: ")
lines_list = read_file_to_list(file_name)

print("Lines stored in list:")
print(lines_list)

```

```python
# 16. Write a Python program to read a file line by line and store it into a variable.
def read_file_to_variable(file_name):
    try:
        content = ""
        with open(file_name, 'r') as file:
            for line in file:
                content += line
        return content
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
        return ""
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")
        return ""

# Example usage
file_name = input("Enter the file name: ")
file_content = read_file_to_variable(file_name)

print("File content stored in variable:")
print(file_content)
```

```python
# 18. Write a Python program to find the longest words.
def find_longest_words(file_name):
    try:
        with open(file_name, 'r') as file:
            words = file.read().split()
        
        max_length = max(len(word) for word in words)
        longest_words = [word for word in words if len(word) == max_length]

        print(f"Longest word length: {max_length}")
        print("Longest word(s):", longest_words)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")
    except ValueError:
        print("Error: The file is empty.")

# Example usage
file_name = input("Enter the file name: ")
find_longest_words(file_name)
```

```python
# 20. Write a Python program to count the frequency of words in a file.
from collections import Counter
import string

def count_word_frequency(file_name):
    try:
        with open(file_name, 'r') as file:
            text = file.read().lower()

        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Split into words
        words = text.split()

        # Count frequency using Counter
        word_freq = Counter(words)

        print("Word Frequency:")
        for word, freq in word_freq.items():
            print(f"{word}: {freq}")
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")

# Example usage
file_name = input("Enter the file name: ")
count_word_frequency(file_name)


```

```python
# 23. Write a Python program to copy the contents of a file to another file.
def copy_file(source_file, destination_file):
    try:
        with open(source_file, 'r') as src:
            content = src.read()
        with open(destination_file, 'w') as dest:
            dest.write(content)
        print(f"Contents copied from '{source_file}' to '{destination_file}'.")
    except FileNotFoundError:
        print(f"Error: The file '{source_file}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file operation.")

# Example usage
source = input("Enter the source file name: ")
destination = input("Enter the destination file name: ")
copy_file(source, destination)
```

```python
# 24. Write a Python program to combine each line from the first file with the corresponding line in the second file.
def combine_lines(file1, file2):
    try:
        with open(file1, 'r') as f1, open(file2, 'r') as f2:
            for line1, line2 in zip(f1, f2):
                combined = line1.strip() + " " + line2.strip()
                print(combined)
    except FileNotFoundError:
        print("Error: One or both files were not found.")
    except PermissionError:
        print("Error: Permission denied for one of the files.")

# Example usage
file1 = input("Enter the first file name: ")
file2 = input("Enter the second file name: ")
combine_lines(file1, file2)
```

```python
# 25. Write a Python program to read a random line from a file.
import random

def read_random_line(file_name):
    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()
            if not lines:
                print("The file is empty.")
                return
            random_line = random.choice(lines).strip()
            print(f"Random line: {random_line}")
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")

# Example usage
file_name = input("Enter the file name: ")
read_random_line(file_name)

```

```python
# 29. Write a Python program to extract characters from various text files and put them into a list.
def extract_characters_from_files(file_list):
    characters = []
    try:
        for file_name in file_list:
            with open(file_name, 'r') as file:
                content = file.read()
                characters.extend(list(content))  # Add each character to the list
        return characters
    except FileNotFoundError as e:
        print(f"Error: The file '{e.filename}' was not found.")
    except PermissionError as e:
        print(f"Error: Permission denied for file '{e.filename}'.")

# Example usage
files = input("Enter file names separated by commas: ").split(',')
files = [file.strip() for file in files]  # Remove extra spaces
chars = extract_characters_from_files(files)

if chars:
    print("Extracted characters:")
    print(chars)
```

```python
# 30. Write a Python program to generate 26 text files named A.txt, B.txt, and so on up to Z.txt.
import string
import os

def create_alphabet_files(directory="."):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)

        for letter in string.ascii_uppercase:
            file_name = os.path.join(directory, f"{letter}.txt")
            with open(file_name, 'w') as file:
                file.write(f"This is file {letter}.txt\n")
        print(f"26 files (A.txt to Z.txt) created in '{directory}'")
    except PermissionError:
        print("Error: Permission denied to create files.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
create_alphabet_files("alphabet_files")  # Creates files in 'alphabet_files' folder
```

```python
# 31. Write a Python program to create a file where all letters of the English alphabet are listed with a specified number of letters on each line.
import string

def write_alphabet_to_file(file_name, letters_per_line):
    try:
        alphabet = string.ascii_uppercase  # 'A' to 'Z'
        with open(file_name, 'w') as file:
            for i in range(0, len(alphabet), letters_per_line):
                line = alphabet[i:i + letters_per_line]
                file.write(line + "\n")
        print(f"Alphabet written to '{file_name}' with {letters_per_line} letters per line.")
    except ValueError:
        print("Error: letters_per_line must be a positive integer.")
    except PermissionError:
        print(f"Error: Permission denied for file '{file_name}'.")

# Example usage
file_name = input("Enter file name: ")
n = int(input("Enter number of letters per line: "))
write_alphabet_to_file(file_name, n)
```
