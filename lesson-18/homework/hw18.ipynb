```python
import pandas as pd
df = pd.read_csv('data\\tackoverflow_qa.csv')
df.head()
```

```python
titanic_df = pd.read_csv("data\\titanic.csv")
titanic_df.head()
```

```python
# Find all questions that were created before 2014
# Find all questions with a score more than 50
# Find all questions with a score between 50 and 100
# Find all questions answered by Scott Boston
# Find all questions answered by the following 5 users
# Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
# Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
# Find all questions that are not answered by Scott Boston
```

```python
before_2014 = df[df['creationdate'] < "2014-01-01"]
print(before_2014)
```

```python
high_score = df[df['score'] > 50]

# Show results
print(high_score)
```

```python
# Find all questions with a score between 50 and 100
score_range = df[df['score'].between(50, 100)]

# Show results
print(score_range)
```

```python
# Find all questions answered by Scott Boston
scott_answers = df[df['ans_name'] == "Scott Boston"]

# Show results
print(scott_answers)
```

```python
# Find all questions answered by the following 5 users
# List of 5 users
users = ["Scott Boston", "unutbu", "jezrael", "Warren Weckesser", "Zero"]  

# Filter questions answered by these users
answered_by_users = df[df['ans_name'].isin(users)]

# Show results
print(answered_by_users)
```

```python
# Find all questions that were created between March, 
# 2014 and October 2014 that were answered by Unutbu and have score less than 5.
# Convert creationdate to datetime
df['creationdate'] = pd.to_datetime(df['creationdate'])

# Apply filters
filtered = df[
    (df['creationdate'] >= "2014-03-01") &
    (df['creationdate'] <= "2014-10-31") &
    (df['ans_name'] == "unutbu") &
    (df['score'] < 5)
]

# Show results
print(filtered)
```

```python
# Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
# Apply filters
filtered = df[
    (df['score'].between(5, 10)) |
    (df['viewcount'] > 10000)
]

# Show results
print(filtered)

# If you just want ID, score, viewcount, and title
print(filtered[['id', 'score', 'viewcount', 'title']])
```

```python
# Find all questions that are not answered by Scott Boston
# Filter questions not answered by Scott Boston
not_scott = df[df['ans_name'] != "Scott Boston"]

# Show results
print(not_scott)

# If you just want ID, title, and answerer's name
print(not_scott[['id', 'title', 'ans_name']])
```

```python
# Select Female Passengers in Class 1 with Ages between 20 and 30: 
# Extract a DataFrame containing female passengers in Class 1 with ages between 20 and 30.

# Load dataset (replace with your actual file name)
# titanic_df = pd.read_csv("titanic.csv")

# Filter female passengers in Class 1 with ages between 20 and 30
female_class1_20_30 = titanic_df[
    (titanic_df['Sex'] == "female") &
    (titanic_df['Pclass'] == 1) &
    (titanic_df['Age'].between(20, 30))
]

# Show results
print(female_class1_20_30)

# If you just want a few columns
print(female_class1_20_30[['Name', 'Sex', 'Age', 'Pclass']])
```

```python
# Filter Passengers Who Paid More than $100: Create a DataFrame with passengers who paid a fare greater than $100.
# Filter passengers with fare > 100
fare_over_100 = titanic_df[titanic_df['Fare'] > 100]

# Show results
print(fare_over_100)

# If you just want selected columns
print(fare_over_100[['Name', 'Sex', 'Age', 'Pclass', 'Fare']])
```

```python
# Select Passengers Who Survived and Were Alone: Filter passengers who survived 
# and were traveling alone (no siblings, spouses, parents, or children).

# Filter passengers who survived and were traveling alone
survived_alone = titanic_df[
    (titanic_df['Survived'] == 1) &
    (titanic_df['SibSp'] == 0) &
    (titanic_df['Parch'] == 0)
]

# Show results
print(survived_alone)

# If you just want selected columns
print(survived_alone[['Name', 'Sex', 'Age', 'Pclass', 'Survived']])
```

```python
# Filter Passengers Embarked from 'C' and Paid More Than $50: Create a DataFrame 
# with passengers who embarked from 'C' and paid more than $50.
# # Load dataset
# titanic_df = pd.read_csv("titanic.csv")

# Filter passengers embarked from 'C' and paid more than $50
embarked_c_over_50 = titanic_df[
    (titanic_df['Embarked'] == 'C') &
    (titanic_df['Fare'] > 50)
]

# Show results
print(embarked_c_over_50)

# If you just want selected columns
print(embarked_c_over_50[['Name', 'Sex', 'Age', 'Pclass', 'Embarked', 'Fare']])
```

```python
# Select Passengers with Siblings or Spouses and Parents or Children: 
# Extract passengers who had both siblings or 
# spouses aboard and parents or children aboard.
# Filter passengers with both siblings/spouses and parents/children aboard
with_family = titanic_df[
    (titanic_df['SibSp'] > 0) &
    (titanic_df['Parch'] > 0)
]

# Show results
print(with_family)

# If you just want selected columns
print(with_family[['Name', 'Sex', 'Age', 'Pclass', 'SibSp', 'Parch']])
```

```python
# Filter Passengers Aged 15 or Younger Who Didn't Survive: 
# Create a DataFrame with passengers aged 15 or younger who did not survive.
# Filter passengers aged 15 or younger who did not survive
young_not_survived = titanic_df[
    (titanic_df['Age'] <= 15) &
    (titanic_df['Survived'] == 0)
]

# Show results
print(young_not_survived)

# If you just want selected columns
print(young_not_survived[['Name', 'Sex', 'Age', 'Pclass', 'Survived']])
```

```python
# Select Passengers with Cabins and Fare Greater Than $200: 
# Extract passengers with known cabin numbers and a fare greater than $200.

# Filter passengers with a cabin and fare greater than $200
cabin_fare_over_200 = titanic_df[
    (titanic_df['Cabin'].notna()) &
    (titanic_df['Fare'] > 200)
]

# Show results
print(cabin_fare_over_200)

# If you just want selected columns
print(cabin_fare_over_200[['Name', 'Sex', 'Age', 'Pclass', 'Cabin', 'Fare']])
```

```python
# Filter Passengers with Odd-Numbered Passenger IDs:
# Create a DataFrame with passengers whose PassengerId is an odd number.
# Filter passengers with odd PassengerId
odd_passenger_ids = titanic_df[titanic_df['PassengerId'] % 2 != 0]

# Show results
print(odd_passenger_ids)

# If you just want selected columns
print(odd_passenger_ids[['PassengerId', 'Name', 'Sex', 'Age', 'Pclass']])
```

```python
# Select Passengers with Unique Ticket Numbers: 
# Extract a DataFrame with passengers having unique ticket numbers.
# Filter passengers with unique ticket numbers
unique_tickets = titanic_df[~titanic_df['Ticket'].duplicated(keep=False)]

# Show results
print(unique_tickets)

# If you just want selected columns
print(unique_tickets[['PassengerId', 'Name', 'Ticket', 'Fare', 'Pclass']])
```

```python
# Filter Passengers with 'Miss' in Their Name and Were in Class 1: Create a DataFrame with female passengers having 
# 'Miss' in their name and were in Class 1.
# Filter female passengers with 'Miss' in their name and in Class 1
miss_class1 = titanic_df[
    (titanic_df['Sex'] == 'female') &
    (titanic_df['Name'].str.contains("Miss")) &
    (titanic_df['Pclass'] == 1)
]

# Show results
print(miss_class1)

# If you just want selected columns
print(miss_class1[['PassengerId', 'Name', 'Sex', 'Age', 'Pclass']])
```
